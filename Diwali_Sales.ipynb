Exploratory Data Analysis

1. Gender


```python
ax = sns.countplot( x='Gender', data = dataset)
sns.set(rc={'figure.figsize':(6,4)})

for bars in ax.containers:          # This is  used to display the numeric amount of the data 
    ax.bar_label(bars)
    
```

```python
dataset.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values( by='Amount', ascending=False)
```

To view this data in the form of graph, we use Bar graph

```python
sales_gender = dataset.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values( by='Amount', ascending=False)

sns.barplot( x='Gender', y='Amount', data=sales_gender)
```

From the above Graph, we can simply conclude, that the number of buyers are greater in women as compared to men.

2. Age

```python
sns.countplot(x='Age Group', hue='Gender', data=dataset)
```

```python
ax = sns.countplot(x='Age Group', hue='Gender', data=dataset)

for bars in ax.containers:
    ax.bar_label(bars)
```

In this graph, the values are return along with the bars


```python
sales_age = dataset.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values( by='Amount', ascending=False)
sns.barplot( x='Age Group', y='Amount', data=sales_age,  hue='Age Group')
```

From the above graph, we can conclude that most of the buyers are of age group between 26-35 yrs Female

3. State

```python
sales_state = dataset.groupby(['State'], as_index=False)['Orders'].sum().sort_values( by='Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(16,6)})
sns.barplot( x='State', y='Orders', data=sales_state,  hue='State')
plt.title('Orders vs States')
```

head() is used to retrive top 10 states only

```python
sales_state = dataset.groupby(['State'], as_index=False)['Amount'].sum().sort_values( by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(16,6)})
sns.barplot( x='State', y='Amount', data=sales_state,  hue='State')
plt.title('Amounts vs States')
```

Here, Kerala has more orders but there purchasing power is low as compared to haryana, bihar, and gujrat.

From the above graph we can conclude that most of the Orders and Total amount spend are from Uttar Pradesh, Maharashtra and Karnataka

4.Martial Status

```python
ax = sns.countplot(x='Marital_Status' , data=dataset , hue='Marital_Status')

sns.set(rc={'figure.figsize':(6,4)})
for bars in ax.containers:
    ax.bar_label(bars)
plt.xlabel('Marital Status')
```

```python
sales_status = dataset.groupby(['Marital_Status','Gender'], as_index=False) ['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(7,5)})
sns.barplot(x='Marital_Status', y='Amount', data=sales_status, hue='Gender')
plt.xlabel('Marital Status')
plt.title('Amount vs Marital Status')
```

From the above graph, we can conclude that most of the buyers are married (women).

5. Occupation

```python
ax = sns.countplot( x='Occupation' , data=dataset, hue='Occupation')
sns.set(rc={'figure.figsize':(21,8)})

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_occu = dataset.groupby(['Occupation'],  as_index=False) ['Amount'].sum().sort_values( by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(20,6) , 'axes.facecolor':'White'})
sns.barplot( x='Occupation', y='Amount', data=sales_occu, hue='Occupation')
```

From the above graph we see that most of the buyers are working in IT sector, Healthcare and Aviation.

6. Product Category

```python
dataset.columns
```

```python
ax = sns.countplot(x='Product_Category', data=dataset, hue='Product_Category')
sns.set(rc={'figure.figsize':(25,10)})
plt.xlabel('Product Categories')
plt.xticks(rotation=55)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_product = dataset.groupby(['Product_Category'], as_index=False) ['Amount'].sum().sort_values( by='Amount', ascending=False).head(10)
sns.barplot(x='Product_Category', y='Amount', data=sales_product, hue='Product_Category')
sns.set(rc={'figure.figsize':(20,8)})
plt.xlabel('Products')
plt.title('Products vs Amounts')
```

Here, The sales of orders of Clothing is More than Food. But the Amount spend on food is comparatively higher than on Clothing.
